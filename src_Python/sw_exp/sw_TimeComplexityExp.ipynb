```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = [5,10,50,100,500,1000,5000,10000,50000,100000]
f = open("Tn.txt",'r')
y1 = []
y2 = []
y3 = []
y4 = []

for i in range(0,10):
    data = float(f.readline())
    y1 = np.append(y1,data)
    
for i in range(10,20):
    data = float(f.readline())
    y2 = np.append(y2,data)

for i in range(20,30):
    data = float(f.readline())
    y3 = np.append(y3,data)
    
for i in range(30,40):
    data = float(f.readline())
    y4 = np.append(y4,data)
    
f.close()

#plt.figure(1)
#plt.title('Insertion sort')
#plt.plot(x, y1)
#plt.figure(2)
#plt.title('Bubble sort')
#plt.plot(x, y2)
#plt.figure(3)
#plt.title('Quick sort')
#plt.plot(x, y3)
#plt.figure(4)
#plt.title('Merge sort')
#plt.plot(x, y4)

#plt.figure(5)
plt.title('Time complexity analysis')
plt.xlabel('Input size')
plt.ylabel('Running time[microsecond]')
plt.semilogy(x, y1, label='Insertion')
plt.semilogy(x, y2, label='Bubble')
plt.semilogy(x, y3, label='Qucik')
plt.semilogy(x, y4, label='Merge')
plt.legend()


```
